```python
import numpy as np;

def bisection(f, es, a, b):
    iter = 0;
    while(1):
        p = a + (b-a)/2;
        if f(p) == 0 or (abs(f(p)) <= es):
            break;

        if f(a) * f(p) < 0:
            b = p;
        else:
            a = p;
        iter += 1;
    print("total iter: {}".format(iter));            
    return p;

def a(x):
    return x - 2**(-x);

def b(x):
    return np.exp(x) - x**2 + 3*x - 2;

def c(x):
    return 2*x*np.cos(2*x) - (x+1)**2;

def d(x):
    return x*np.cos(x) - 2*x**2 + 3*x - 1;

print("5.a")
p = bisection(a, 10**(-5), 0, 1);
print("approximate x: {}, y:{}".format(p,a(p)));

print("5.b")
p = bisection(b, 10**(-5), 0, 1);
print("approximate x: {}, y:{}".format(p,b(p)));

print("5.c")
p1 = bisection(c, 10**(-5), -3, -2); p2 = bisection(c, 10**(-5), -1, 0);
print("approximate x: {}, y:{}".format([p1, p2], [c(p1),c(p2)]));

print("5.d")
p1 = bisection(d, 10**(-5), 0.2, 0.3); p2 = bisection(d, 10**(-5), 1.2, 1.3);
print("approximate x: {}, y:{}".format([p1, p2], [d(p1),d(p2)]));
```
