```python
import numpy as np;
import matplotlib.pyplot as plt;
```

```python
a = lambda x: x**2-2*x+3
b = lambda x: x**3
c = lambda x: 1/(x+2)
d = lambda x: np.exp(x)
e = lambda x: 1/2*np.cos(x)+1/3*np.sin(2*x)
f = lambda x: np.log(x+2)

a1 = lambda x: 10/3-2*x
b1 = lambda x: 0+3/5*x;
c1 = lambda x: np.log(3)/2+(3-3*np.log(3))*x;
d1 = lambda x: (np.exp(2)-1)/(2*np.exp(1))+(3/np.exp(1))*x;
e1 = lambda x: (np.sin(1)/2)+(np.sin(2)/4-np.cos(2)/2)*x;
f1 = lambda x: (3*np.log(3)-2)/2 + (3-9*np.log(3)/4)*x;

a2 = lambda x: 3+(-2*x)+1*x**2
b2 = lambda x: 0+(3/5)*x+0*x**2
c2 = lambda x: (-51*np.log(3)+60)/8 + (3-3*np.log(3))*x + (165*np.log(3)-180)/8*x**2
d2 = lambda x: (-3*np.exp(2)+33)/(4*np.exp(1)) + 3/np.exp(1)*x + (15*np.exp(2)-105)/(4*np.exp(1))*x**2
e2 = lambda x: (72*np.sin(1)-90*np.cos(1))/24 + (np.sin(2)-2*np.cos(2))/4*x +(90*np.cos(1)-60*np.sin(1))/8*x**2
f2 = lambda x: (-54*np.log(3)+76)/24 + (12-9*np.log(3))/4*x + (90*np.log(3)-100)/8*x**2

interval = np.arange(-1,1,0.01);

for actual,approx1, approx2 in zip([a,b,c,d,e,f],[a1,b1,c1,d1,e1,f1],[a2,b2,c2,d2,e2,f2]):
    plt.plot(interval, actual(interval),'b', label="actual");
    plt.plot(interval, approx2(interval),'g', label="2 degree approx");
    plt.legend()
    plt.show()
```
