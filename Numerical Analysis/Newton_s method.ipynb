```python
import numpy as np;

def d(x):
    return np.sqrt(x**4 + x**2 - 2*x + 1);

def f(x):
    return 4 * x**3 + 2*x - 2;

def df(x):
    return 12 * x**2 + 2;

def newton(f, df, p0, tol, maxit=1000):
    i = 1;
    while(i <= maxit):
        if df(p0) == 0:
            print("df(p0) = 0 : can't be processed by Newton's method!");
            break;
        p = p0 - f(p0)/df(p0);
        print("iter: {}, {}".format(i, p));
        if abs(p-p0) <= tol:
            return p;
        i += 1;
        p0 = p;

p = newton(f, df, 1, 1e-4);
print("p, f(p), d(p): {}, {}, {}".format(p, f(p), d(p)));
print("isMinimum: {}".format(d(p) < d(p + 1e-3) and d(p) < d(p - 1e-3)));
```

```python
import numpy as np;

def f(x):
    return 1/2 + 1/4 * x**2 - x*np.sin(x) -1/2*np.cos(2*x);

def df(x):
    return 1/2*x - np.sin(x) - x * np.cos(x) + np.sin(2*x);

def newton(f, df, p0, tol, maxit=1000):
    i = 1;
    while(i <= maxit):
        if df(p0) == 0:
            print("df(p0) = 0 : can't be processed by Newton's method!");
            break;
        p = p0 - f(p0)/df(p0);
        print("iter: {}, {}".format(i, p));
        if abs(p-p0) <= tol:
            return p;
        i += 1;
        p0 = p;

print("p0: pi/2")
p = newton(f, df, np.pi/2, 1e-5);
print("p, f(p): {}, {}".format(p, f(p)));

print("\np0: 5*pi")
p = newton(f, df, 5 * np.pi, 1e-5);
print("p, f(p): {}, {}".format(p, f(p)));

print("\np0: 10*pi")
p = newton(f, df, 10 * np.pi, 1e-5);
print("p, f(p): {}, {}".format(p, f(p)));
```

```python
import numpy as np;

def a(x):
    fx = 1 - 4 * x * np.cos(x) + 2 * x**2 + np.cos(2*x);
    dx = -2 * np.sin(2*x) + 4*x*np.sin(x) -4*np.cos(x) + 4*x;
    return [fx, dx];

def b(x):
    fx = x**6 + 6*x**5 + 9*x**4 - 2*x**3 - 6*x**2 + 1;
    dx = 6*x**5 + 30*x**4 + 36*x**3 - 6*x**2 - 12*x;
    return [fx, dx];

def c(x):
    fx = np.sin(3*x)+3*np.exp(-2*x)*np.sin(x)-3*np.exp(-x)*np.sin(2*x)-np.exp(-3*x);
    dx = np.exp(-3*x)*(3*np.exp(3*x)*np.cos(3*x) + 3*np.exp(2*x)*np.sin(2*x) - 6*np.exp(2*x)*np.cos(2*x)-6*np.exp(x)*np.sin(x) + 3*np.exp(x)*np.cos(x)+3);
    return [fx, dx];

def d(x):
    fx = np.exp(3*x)-27*x**6+27*x**4*np.exp(x)-9*x**2*np.exp(2*x);
    dx = 3*np.exp(3*x) + (-18*x**2 - 18*x)*np.exp(2*x) +(27*x**4 + 108*x**3)*np.exp(x) - 162*x**5;
    return [fx, dx];

def newton(f, p0, tol, maxit=1000):
    i = 1;
    while(i <= maxit):
        fx, dx = f(p0);
        if dx == 0:
            print("df(p0) = 0 : can't be processed by Newton's method!");
            break;
        p = p0 - fx/dx;
        print("iter: {}, {}".format(i, p));
        if abs(p-p0) <= tol:
            return p;
        i += 1;
        p0 = p;

print("a.")
p = newton(a, 0.5, 1e-5);
print("p, f(p): {}, {}".format(p, a(p)[0]));

print("\nb.")
p = newton(b, -2.5, 1e-5);
print("p, f(p): {}, {}".format(p, b(p)[0]));

print("\nc.")
p = newton(c, 3.5, 1e-5);
print("p, f(p): {}, {}".format(p, c(p)[0]));

print("\nd.")
p = newton(d, 4, 1e-5);
print("p, f(p): {}, {}".format(p, d(p)[0]));
```

```python
import numpy as np;

def f(x):
    fx = np.exp(6*x) + 1.441*np.exp(2*x)-2.079*np.exp(4*x)-0.3330;
    dx = (3000*np.exp(6*x)-4158*np.exp(4*x)+1441*np.exp(2*x))/500;
    d2x = (9000*np.exp(6*x)-8316*np.exp(4*x)+1441*np.exp(2*x))/250;
    return [fx, dx, d2x];

def newton(f, p0, tol, maxit=1000):
    i = 1;
    while(i <= maxit):
        fx, dx, d2x = f(p0);
        if dx == 0:
            print("df(p0) = 0 : can't be processed by Newton's method!");
            break;
        p = p0 - fx/dx;
        print("iter: {}, {}".format(i, p));
        if abs(p-p0) <= tol:
            return p;
        i += 1;
        p0 = p;

def modified_newton(f, p0, tol, maxit=1000):
    i = 1;
    while(i <= maxit):
        fx, dx, d2x = f(p0);
        if dx == 0:
            print("df(p0) = 0 : can't be processed by Newton's method!");
            break;
        p = p0 - (fx*dx)/(dx**2 - fx*d2x);
        print("iter: {}, {}".format(i, p));
        if abs(p-p0) <= tol:
            return p;
        i += 1;
        p0 = p;

print("newton.")
p = newton(f, -0.5, 1e-5);
print("p, f(p): {}, {}".format(p, a(p)[0]));

print("\nmodified_newton.")
p = modified_newton(f, -0.5, 1e-5);
print("p, f(p): {}, {}".format(p, a(p)[0]));
```
